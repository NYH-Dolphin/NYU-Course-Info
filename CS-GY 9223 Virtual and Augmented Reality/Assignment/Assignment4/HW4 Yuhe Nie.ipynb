```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
from PIL import Image, ImageFilter
import math
```

```python
#import the example image
image = io.imread("https://www.printablee.com/postpic/2015/06/free-printable-eye-chart_112469.jpg")
image = cv2.resize(image, (360, 480))
cv2_imshow(image)
```

```python
def circle_of_confusion(f, D, S1, S):
  M = f/(S1-f)
  c = M*D*(abs(S-S1)/S)
  return c
```

```python
#TODO: get the equation parameters from the user
f = 17
D = 20
S1 = 30
S = 1000

#TODO: calulate the circle of confusion
c = circle_of_confusion(f,D,S1,S)
print('circle of confusion: ' + str(c))
```

```python
#calculate the size of the blur
PPI = 96
blur_radius = c * (PPI/25.4) / 2
print('blur radius: ' + str(blur_radius))
```

```python
#apply a gaussian filter with the radius of the blur
filter_size = int(blur_radius)
blurred_img = cv2.GaussianBlur(image, (filter_size, filter_size), blur_radius)
cv2_imshow(blurred_img)
```
